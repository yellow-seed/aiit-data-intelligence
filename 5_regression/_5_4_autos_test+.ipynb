````python
import pandas as pd
import numpy as np
autos = pd.read_csv('data/imports-85.data', na_values='?')
autos.columns = ['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration',
                    'num_of_doors', 'body_style', 'drive_wheels', 'engine_location',
                    'wheel_base', 'length','width', 'height', 'curb_weight',
                    'engine_type', 'num_of_cylinders', 'engine_size', 'fuel_system',
                    'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm',
                    'city_mpg', 'highway_mpg','price']
autos
````

````python
# 表示する最大列数の設定
pd.set_option('display.max_columns', len(autos.columns))
````

````python
# 欠損値 NaN がある行の除去
autos = autos.dropna()
autos
````

# 授業参加確認の問題

imports-85.data の欠損値 NaN がある行の除去した後のデータセットを処理して、以下の質問に答えなさい。
数値は、小数点以下3桁目を四捨五入し、小数点以下2桁まで応えること（1.235ならば1.24が答え）。

2. 4気筒（num_of_cylindersが'four'）の車両の中で最も多いメーカー (make) は何か
3. 最も値段 (price) が高い車両のメーカー (make) は何か
4. make, horsepower を説明変数、price を目的変数として作成した線形重回帰モデルの決定係数はいくつか

````python
# ４気筒（num_of_cylindersが'four'）の車両
autos.query("num_of_cylinders=='four'")
````

````python
# ２．４気筒（num_of_cylindersが'four'）の車両の中で最も多いメーカー (make) は何か
autos.query("num_of_cylinders=='four'").make.value_counts()
````

````python
# ３．最も値段 (price) が高い車両のメーカー (make) は何か
autos[autos.price==autos.price.max()]
````

````python
# ４．make, horsepower を説明変数、price を目的変数として作成した線形重回帰モデルの決定係数はいくつか
# - make はダミー変数化
from sklearn.linear_model import LinearRegression
_data = autos[['make', 'horsepower']]
X = pd.get_dummies(data=_data, drop_first=True)
Y = autos.price

# ダミー変数化を確認
X
````

````python
model = LinearRegression()
model.fit(X, Y)
# 決定係数
from sklearn.metrics import r2_score
Y_predict = model.predict(X)
print('R^2={:.2f}'.format(r2_score(Y, Y_predict)))
````

### 質問：ダミー変数化

上記のようにメーカー (make) はダミー変数化することで回帰モデルの説明変数にすることができる。

### 質問：平均二乗誤差の値が大きい

平均二乗誤差の値の大小はデータに依存。
- 同じデータについて見ている時に、モデルのデータへの適合度合いを見比べる場合には平均二乗誤差も使える

決定係数
- 客観的に（いろいろなデータについて見ている時に）モデルのデータへの適合度合いを見比べる場合には決定係数

````python
from sklearn.metrics import mean_squared_error

print('MSE={:.3f}'.format(mean_squared_error(Y, Y_predict)))
````



### 質問：決定係数 0.92 の意味

```
馬力が高い車を扱っているメーカーは価格が高いということが、決定係数 0.92 からいえるのか？
```

高い決定係数から言えることは、馬力とメーカの2つを説明変数に採用することで、<u>全体としては</u>データ（価格）にあてはまりが良いと見做せるモデルを作成することができた、ということ。

- 全データで horsepower のみを説明変数に用いた単回帰分析

````python
X_all = autos[['horsepower']]
Y_all = autos.price

model_all = LinearRegression()
model_all.fit(X_all, Y_all)

# 決定係数
Y_predict = model_all.predict(X_all)
print('R^2={:.3f}'.format(r2_score(Y_all, Y_predict)))
````

- メーカ別に horsepower のみを説明変数に用いた単回帰分析

````python
model = LinearRegression()
makes = autos.make.unique()
df_r2 = pd.DataFrame(columns=['R2'], index=makes)

for make in makes:
    _data = autos[autos['make']==make]
    X = _data[['horsepower']]
    Y = _data.price

    model.fit(X, Y)

    # 決定係数
    Y_predict = model.predict(X)
    df_r2.at[make] = r2_score(Y, Y_predict)

df_r2.sort_values('R2', ascending=False)
````
