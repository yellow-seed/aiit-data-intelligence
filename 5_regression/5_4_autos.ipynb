````python
import pandas as pd

# 欠損値が '?' で入っていることを na_values='?' で指定
# - 指定しなければ '?' を含む列は全て文字列として読み込まれる
autos = pd.read_csv('data/imports-85.data', na_values='?')

# imports-85.data には列ラベルがついていないため imports-85.names から列ラベルを付ける
autos.columns = ['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration',
                    'num_of_doors', 'body_style', 'drive_wheels', 'engine_location',
                    'wheel_base', 'length','width', 'height', 'curb_weight',
                    'engine_type', 'num_of_cylinders', 'engine_size', 'fuel_system',
                    'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm',
                    'city_mpg', 'highway_mpg','price']

# データの確認
# - 欠損値は NaN と表示
autos
````

````python
# 表示する最大列数の設定
pd.set_option('display.max_columns', len(autos.columns))
# 表示する最大行数の設定
# pd.set_option('display.max_rows', len(autos))
````

````python
# DataFrame の要約を表示
autos.info()
````

````python
autos.isna() # 欠損値かどうか判定
````

````python
# 欠損値 NaN の個数
autos.isna().sum()
````

````python
# 欠損値 NaN を含む行の除去
# - 欠損値を含む行を全て除去
df = autos.dropna()

# 確認
# - 0, 1, 4, 6, 8行目が無くなっている
# - rows が 204 から 159 に減少
df
````

````python
# TODO: 視聴確認用の記述をここにする
# num_of_cylinders == 'four' の中で最も数が多い
df.where(df['num_of_cylinders'] == 'four').groupby('make').size()
````

````python
# price が最大の make
df.loc[df['price'] == df['price'].max()]['make']
# max_price = df['price'].max()
# df['make'].where(df['price'] == df['price'].max())
````

````python
# 線形重回帰モデルの作成
from sklearn.linear_model import LinearRegression
# 説明変数　make, horsepower データフレームからdrop
# X = df.drop(['make', 'horsepower'], axis=1)
X = df[['horsepower']]

# 目的変数 price
Y = df.price

# model作成
model = LinearRegression()
model.fit(X, Y)
````

````python
# 平均二乗誤差、決定係数の出力
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# 予測値Y
Y_predict = model.predict(X)

# MSE
mse = mean_squared_error(Y, Y_predict)
print(mse)

# r2_score
R2_score = r2_score(Y, Y_predict)
print(R2_score)
````
