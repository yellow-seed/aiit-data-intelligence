````python
# 列のラベル
df_math.columns
````

````python
df_math.drop(labels='school', axis=1)
````

````python
# 列の削除
# - axis=1 で列を削除することを指定
# - inplace=True を指定すると、直接データフレームを処理する
#   指定しないと、もとのデータフレームはそのままで、列が削除されたデータフレームが返される
df_math.drop(labels='romantic', axis=1, inplace=True)
df_math.columns
````

````python
# 複数の列の削除
df_math.drop(labels=['Dalc', 'Walc'], axis=1, inplace=True)
df_math.columns
````

````python
# 列のユニークな値
df_math['sex'].unique()
````

````python
# 全ての列のユニークな値
for row in df_math.columns:
    print(row, df_math[row].unique())
````

````python
# 全ての列のユニークな値をソートして出力
for row in df_math.columns:
    val = df_math[row].unique()
    val.sort()
    print(row, val)
````

````python
# 全ての列のユニークな値をdict型変数に保存
# - sort() はもとのリストを破壊的に操作
# - ソートされたリストを戻り値として欲しい場合は sorted()
row_dict = {}

for row in df_math.columns:
    val = df_math[row].unique()
    val.sort()
    row_dict[row] = val

row_dict
````

````python
# G1, G3列のみ取り出して相関を求める
df_math[['G1', 'G3']].corr()
````
